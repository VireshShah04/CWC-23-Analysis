```python
import numpy as np
import pandas as pd
```

```python
# Reading the data file
df = pd.read_csv('CWC23.csv')
```

```python
df.head()
df.tail()
```

```python
# Getting the shape of data
df.shape
```

```python
# Getting the columns
df.columns
```

```python
# Slicing the bowlers data from the csv
df_bowlers = df[df['bat_or_bowl'] == 'bowl']
```

```python
df_bowlers.head()
```

```python
# Dealing with NaN values from bowler df
df_bowlers.dropna(axis = 1, inplace = True)
```

```python
df_bowlers.head()
```

```python
# Checking for duplicate values
df_bowlers.duplicated()
```

```python
# Getting a description of data
df_bowlers.describe()
```

```python
# Renaming some columns
df_bowlers.rename(columns = {'bb_bf': 'Balls Bowled'}, inplace = True)
```

```python
df_bowlers.columns
```

```python
# Creating dataframe of indian bowlers
df_ind_bowlers = df_bowlers[df_bowlers['team'] == 'IND']
```

```python
df_ind_bowlers.head()
```

```python
# Shape of our bowlers data
df_ind_bowlers.shape
```

```python
# Description of data
df_ind_bowlers.describe()
```

```python
# Checking for duplicate values
df_ind_bowlers['player'].duplicated()
```

```python
# Grouping data based on players name
bowling_data = df_ind_bowlers.groupby('player')
```

```python
total_data = bowling_data.sum().reset_index()
total_data
```

```python
avg_data = bowling_data.mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a bar plot of wickets and player
sns.set(style = 'whitegrid')
# Create a bar chart using Seaborn with additional customizations
sns.barplot(x='player', y='wkts', data=total_data)  # Use a different color palette

plt.xlabel('Players')
plt.ylabel('Wickets')
plt.title('Wickets per player')
plt.xticks(rotation=45, ha='right', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Creating a barplot of maidens per bowler
sns.set(style='whitegrid')
sns.barplot(x='player', y='mdns', data=total_data)
plt.xlabel('Players')
plt.ylabel('Maidens')
plt.title('Maidens per player')
plt.xticks(rotation=45, ha='right', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Creating a bar plot of wickets and player
sns.set(style = 'whitegrid')
# Create a bar chart using Seaborn with additional customizations
sns.lineplot(x='wkts', y='runs', data=total_data)  # Use a different color palette

plt.xlabel('Wickets')
plt.ylabel('Runs')
plt.title('Runs v/s Wickets')

plt.tight_layout()
plt.show()
```
